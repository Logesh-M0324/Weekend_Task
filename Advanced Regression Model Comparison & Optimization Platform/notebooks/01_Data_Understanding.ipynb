```python
for i in numerical_columns_Train_data.corr():
    print(i)
```

```python
corr_matrix = numerical_columns_Train_data.corr()

corr_matrix.items
```

```python
for col_name, row_values in corr_matrix.items():
    for row_name, corr_val in row_values.items():
        
        if col_name != row_name:
            print(f"Correlation between {col_name} and {row_name}: {corr_val:.2f}")
```

# I set My Own Threshold Value of 0.50

```python
threshold = 0.50
```

```python
target_var_corr.__class__
```

```python
relevant_features = {}

for row, cols in target_var_corr.items():
    if cols >= threshold and row != "SalePrice":
        relevant_features[row] = cols
```

```python
relevant_features 
```

```python
plt.Figure(figsize=(12,10))

sns.histplot(relevant_features, kde = True)

plt.title("Relevant features with Sales Price")

plt.show()
```
