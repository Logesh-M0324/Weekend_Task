```python
# Step 1 – Import Required Libraries

import pandas as pd
import numpy as np
```

```python
# Step 2 – Load the Final Dataset

master_df = pd.read_csv(
    "../exports/feature_engineered_dataset.csv",
    parse_dates=[
        "order_purchase_timestamp",
        "order_approved_at",
        "order_delivered_carrier_date",
        "order_delivered_customer_date",
        "order_estimated_delivery_date"
    ]
)
```

```python
# 5.1 Create Project Statistics

# Create a dictionary to store all project metrics.

project_summary = {
    "Total Raw Tables": 9,
    "Tables Merged": 8,
    "Final Dataset Name": "master_df",
    "Final Records": len(master_df),
    "Final Columns": len(master_df.columns)
}

project_summary
```

```python
# 5.2 Missing Value Summary

# If you tracked missing values during Phase 2 (recommended), use those counts. Otherwise, summarize the final dataset:

missing_summary = pd.DataFrame({
    "Column": master_df.columns,
    "Missing Values": master_df.isnull().sum().values,
    "Missing Percentage": (
        master_df.isnull().sum() / len(master_df) * 100
    ).round(2)
})

missing_summary.head()
```

```python
# Export Missing Value Report

missing_summary.to_csv(
    "../exports/missing_value_report.csv",
    index=False
)
```

```python
# 5.3 Duplicate Record Summary

duplicate_summary = pd.DataFrame({
    "Metric": [
        "Duplicate Rows"
    ],
    "Value": [
        master_df.duplicated().sum()
    ]
})

duplicate_summary
```

```python
# Export Duplicate Report

duplicate_summary.to_csv(
    "../exports/duplicate_report.csv",
    index=False
)
```

```python
# 5.4 Outlier Summary

# Tracked outliers during Phase 2, summarize those values here. If not, generate a report for key numerical columns:

numeric_columns = [
    "price",
    "freight_value",
    "payment_value",
    "total_order_value",
    "customer_lifetime_value",
    "total_delivery_days"
]

outlier_summary = []

for column in numeric_columns:

    Q1 = master_df[column].quantile(0.25)
    Q3 = master_df[column].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = (
        (master_df[column] < lower) |
        (master_df[column] > upper)
    ).sum()

    outlier_summary.append(
        {
            "Column": column,
            "Outliers": outliers
        }
    )

outlier_summary = pd.DataFrame(outlier_summary)

outlier_summary
```

```python
# Export Outlier Report

outlier_summary.to_csv(
    "../exports/outlier_report.csv",
    index=False
)
```

```python
# 5.5 Feature Engineering Summary

feature_summary = pd.DataFrame({
    "Feature Created": [
        "order_processing_days",
        "shipping_days",
        "total_delivery_days",
        "delivery_delay_days",
        "on_time_delivery",
        "total_order_value",
        "customer_purchase_count",
        "customer_lifetime_value",
        "repeat_customer",
        "seller_revenue",
        "seller_performance_score",
        "average_review_score"
    ]
})

feature_summary
```
